### Discussion

From the phase diagram we see that when T and P are low, cooperation is favored with more of the players wating until the last round to defect, or not defecting at all. As T and P increases, se move into regime 2 the degree of cooperation decreases, with more agents choosing to defect earlier in the game. Eventually, with large enough T and P, we reach regime 3, where most players defects already in the first round.

These tendencies can be understood by looking at the payoff-matrix:

|     | C    | D     |
| ---:|:----:| :-----:|
| C   | Both get 1 points | Defector gets T points, cooperator gets 0 points|
| D   | Defector gets T points, cooperator gets 0 points| Both get P points|

If T and P is small  (T close to 1, P close to 0), there is no gain in defecting.

And if T and P is large (T close to 2, O close to 1), there is a lot to gain from defecting.
